```python
import pandas as pd
import numpy as np
from pycaret.time_series import *
```

```python
path = '../data/'
```

```python
train_df = pd.read_csv(path + 'preproc3_train.csv')
```

```python
for i in range(1, 101):
    setup_rgs = setup(data=train_df.loc[train_df.building_id == i].drop('building_id', axis=1),
                    target='consumption',
                    index='Time',
                    session_id=0,
                    #   transform_target='box-cox', # 'log', 'sqrt', 'exp', 'cos'
                    #   transform_exogenous='box_cox',
                    seasonal_period='H',
                    fh=24*7,
                    # use_gpu=True
                      )
    top5 = compare_models(sort='SMAPE', n_select=5)
    save_model(top5, path + f"pycaret_model/{i}")
    break
```

```python
from sktime.forecasting.arima import AutoARIMA

tunes=[]
finals = []
for i in range(1, 101):
    setup(data=train_df.loc[train_df.building_id == i].drop(['building_id', 'Solar radiation'], axis=1),
                    target='consumption',
                    index='Time',
                    session_id=0,
                    #   transform_target='box-cox', # 'log', 'sqrt', 'exp', 'cos'
                    #   transform_exogenous='box_cox',
                    seasonal_period='H',
                    fh=24*7,
                    use_gpu=True
                    )
    # model = create_model('auto_arima')
    model = AutoARIMA(sp=24*7, random_state=0, suppress_warnings=True, n_jobs=-1)  
    tuned = tune_model(model, optimize='smape')
    final = finalize_model(estimator=model)
    save_model(final, 
               f"pycaret_model/auto_arima/{i}",
               verbose=False)
    tunes.append(tuned)
    finals.append(final)

    
```
