```python
import json

file_path = "./BWHE217000025083.json"

# 기존 json 파일 읽어오기
with open(file_path, 'r') as file:
    data = json.load(file)['named_entity']
text_list = []
for d in data:
    text_list.append({'text' : d['title'][0]['sentence']})
    for t in d['content']:
        text_list.append({'text' : t['sentence'].replace('..', '').strip()})
```

```python
len(text_list), text_list
```

```python
# 기존 json 파일 덮어쓰기
with open("test_data.json", 'w', encoding='utf-8') as file:
    json.dump(text_list, file, indent="\t")
```
