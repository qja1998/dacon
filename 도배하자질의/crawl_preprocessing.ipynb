```python
import pandas as pd
```

```python
df = pd.read_csv("./data/add_data/dictionary.csv")
df
```

```python
new_list = []
for i in range(len(df)):
    item, content = df.iloc[i]
    new_list.append(f"{item}의 정의는 다음과 같습니다. {content}")

new_df = pd.DataFrame(new_list, columns=['text'])
new_df
```

```python
new_df.to_csv("./data/add_data/preproc_dictionary.csv", index=False)
```
