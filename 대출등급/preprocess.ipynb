```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/train.csv")
test_df = pd.read_csv("data/test.csv")
```

```python
categorical_cols = ['주택소유상태', '대출목적']
```

```python
def preprocessing(df, is_train=True):
    for term in df['근로기간'].unique():
        num = term[:2].strip()
        if num.isdigit():
            df.loc[df['근로기간'] == term, '근로기간'] = int(num)
        elif num == '<' or num == '<1':
            df.loc[df['근로기간'] == term, '근로기간'] = 0
        else:
            df.loc[df['근로기간'] == term, '근로기간'] = -1

    df['근로기간'] = df['근로기간'].astype(int)
    df['근로기간']

    if is_train:
        for grade in df['대출등급'].unique():
            df.loc[df['대출등급'] == grade, '대출등급'] = ord(grade) - 65

    replace_dict = {
        '대출기간' : {' 36 months' : 36, ' 60 months' : 60},
        '주택소유상태' : {'RENT' : 0, 'MORTGAGE' : 2, 'OWN' : 1, 'ANY' : -1},
        '대출목적' : {k : v for v, k in enumerate(['신용 카드', '자동차', '주택 개선', '주요 구매', '부채 통합', '휴가', '의료', '기타', '이사', '결혼', '재생 에너지', '소규모 사업', '주택'])}
    }

    for col in replace_dict:
        value_dict = replace_dict[col]
        for value in value_dict:
            df.loc[df[col] == value, col] = value_dict[value]
    
    return df
```

```python
df = preprocessing(df)
test_df = preprocessing(test_df, False)
```

```python
df.to_csv('./data/train_preproc.csv', index=False)
pd.read_csv('./data/train_preproc.csv')
```

```python
test_df.to_csv('./data/test_preproc.csv', index=False)
pd.read_csv('./data/test_preproc.csv')
```
